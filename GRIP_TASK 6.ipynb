# SPLITING THE DATA INTO DEPENDENT AND INDEPENDENT VARIABLES

```python
from sklearn.model_selection import train_test_split
x = iris.data
y = iris.target
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=42)
```

# IMPORTING THE ALGORTHIM

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
```

```python
d_tree = DecisionTreeClassifier(random_state=42)

d_tree = d_tree.fit(x_train, y_train)
```

# At conclusion,  we can now feed any new/test data to this classifer and it would be able to predict the right class accordingly.

```python
y_pred = d_tree.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.model_selection import cross_val_score # for cross validation
from sklearn.metrics import confusion_matrix # to create a confusion matrix
from sklearn.metrics import classification_report
from sklearn.metrics import plot_confusion_matrix
```

```python
print(confusion_matrix(y_test, y_pred))

print(classification_report(y_test, y_pred))
```
